## Exploring the Data

Now we will read in the data for the `Posts` table.

```python
# Import libraries
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning) 
import pandas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# Import data and store `CreationDate` as a datetime object
questions = pd.read_csv('QueryResults.csv', parse_dates = ['CreationDate'] )
```

The `FavoriteCount` column does have missing values.  This is probably due to the fact that the question was not present in any users' favorite list, so we will replace the missing values with zeroes.  After that, we will convert the type to integer as well, just as most of the other columns are.  The `Tags` column is an object type, so let's delve deeper to see what specific type of object.

```python
questions['Tags'].apply(lambda object:  type(object)).unique()
```

All of the objects in the column are strings.  Also, each question can have a maximum of five tags.  We will just keep the tags as a list.  

## Cleaning the Data

We will commence by cleaning the `FavoriteCount` column.  We will fill in the missing values with zeroes and change the type to an integer. 

```python
questions.fillna(value={'FavoriteCount': 0}, inplace=True)
questions['FavoriteCount'] = questions['FavoriteCount'].astype(int)
questions.dtypes
```

Now we will clean up the `Tags` column by including commas between different tages.

```python
questions["Tags"] = questions["Tags"].str.replace("^<|>$", "").str.split("><")
questions.sample(3)
```

## Most Popular Tags

We will determine which tags are the most popular by first counting how many times each tag was used:

```python
# Count how many times a tag was used via a dictionary
tag_count = {}

for tags in questions['Tags']:
    for tag in tags:
        if tag in tag_count:
            tag_count[tag] += 1
        else:
            tag_count[tag] = 1
            
# Transform dictionary to a dataframe
tag_count = pd.DataFrame.from_dict(tag_count, orient = 'index')
tag_count.rename(columns = {0: 'tag_count'}, inplace = True)
tag_count.head(10)
```

To better understand this data, we will sort the dataframe by `tag_count` and visualize the top 20 results.

```python
# Sort data by most `tag_count`
most_used = tag_count.sort_values(by='tag_count').tail(20)
most_used
```

We can see that machine-learning has a tag count of 2128 while machine-learning-model has a tag count of 172.  This shows us that the tag count decreases rapidly in these 20 results, so the top 20 should be sufficient for our analysis. 

Next we will visualize the data.

```python
# Create a bar graph of the `tag_count` results
most_used.plot(kind = 'barh', figsize = (15, 9))
plt.title('Most Used Tags by Tag Count')
```

Some of the tags are very broad, such as `python`, `r`, and `dataset`, so they will not be useful in our analysis.  Before we delve deeper into analyzing this bar graph, let's repeat the process for views.

```python
# Count how many times a tag was viewed via a dictionary
tag_view_count = {}

for index, row in questions.iterrows():
    for tag in row['Tags']:
        if tag in tag_view_count:
            tag_view_count[tag] += row['ViewCount']
        else:
            tag_view_count[tag] = row['ViewCount']

# Convert dictionary to a dataframe
tag_view_count = pd.DataFrame.from_dict(tag_view_count, orient = 'index')
tag_view_count.rename(columns={0: 'view_count'}, inplace =True)

most_viewed = tag_view_count.sort_values(by = 'view_count').tail(20)

# Create a bar graph of 'ViewCount'
most_viewed.plot(kind = 'barh', figsize = (15, 9))
plt.title('Most Viewed Tags by View Count')
```

It will be much easier to analyze these bar graphs side by side.

```python
# Place bar graphs side by side
fig, axes = plt.subplots(nrows = 1, ncols = 2)
fig.set_size_inches((20, 10))
most_used.plot(kind = 'barh', ax = axes[0], subplots = True)
most_viewed.plot(kind = 'barh', ax = axes[1], subplots = True)
```

Looking at the graphs, it is apparent that machine learning is very popular among the tag count and view count.  Python is such a vast topic in data science, so we will not focus on it.  Deep learning, which is a type of deep learning based on artificial neural networks, is also popular, along with neural network, keras, classification, tensorflow, and scikit-learn.  Tensorflow is a library for deep learning and scikit-learn is one for more general machine learning, including classification.  Thus, deep learning appears to be the most popular topic in data science.

## Connections Between Tags

We will now analyze the relationships between pairs of tags by counting how many times each pair appears together.  We will do this by first creating a list of all tags. 

```python
tags_list = list(tag_count.index)
```

Next we'll organize the list into a dataframe.  The rows and columns will both have tags in them, so each cross-section will be a pair of tags. 

```python
connections = pd.DataFrame(index = tags_list, columns = tags_list)
connections.iloc[0:5, 0:5]
```

In order to populate this dataframe, we will first fill in all values with zeroes.  Then for each list of tags in questions['Tags'], we will increment the intervening tags one by one.  The result will be a dataframe that indicates how many times pairs of tags appeared together.

```python
# Fill dataframe with zeroes
connections.fillna(0, inplace = True)

# Populate dataframe with how many times pairs of tags appear together
for tags in questions['Tags']:
    connections.loc[tags, tags] +=1
```

Since this dataframe is very large, we will just focus on the tags that were used most often.  

```python
# Focus in on most used tags
connections_most_used = connections.loc[most_used.index, most_used.index]

# Function to design cells
def design_cells(x):
    support_df = pd.DataFrame('', index=x.index, columns=x.columns)
    support_df.loc["regression", "predictive-modeling"] = "background-color: yellow"
    support_df.loc["predictive-modeling", "regression"] = "background-color: yellow"
    for k in range(support_df.shape[0]):
        support_df.iloc[k,k] = "color: red"
    
    return support_df

# Apply design to cells
connections_most_used.style.apply(design_cells, axis=None)
```

If we examine the cells that are highlighted in yellow, they represent the amount of times the pair `regression` and `predictive-modeling` are seen together, which is 23 times.  The cells that are in red represent how many times each tag was used.  

In order to visualize this dataframe better, we will create a heatmap.  We will remove the red values since they do not represent pairs and will cause our colors to be skewed. 

```python
# Remove red values
for i in range(connections_most_used.shape[0]):
    connections_most_used.iloc[i, i] = pd.np.NaN
```

```python
# Create a heatmap
plt.figure(figsize = (15, 10))
sns.heatmap(connections_most_used, cmap = 'Purples', annot = False)
plt.title('Heatmap of How Many Times Pairs of Tags are Used Together')
```

The darkest regions of the heatmap indicate pairs of tags that are used most often, and they occur at the bottom right of this heatmap.  These include `machine-learning` and `python`, `machine-learning` and `deep learning`, and `machine-learning` and `neural-network`.  These are also the tags that are used most often individually, so the fact that they occur often in pairs might just be due to the fact that each of them individually is used frequently.  Another shortcoming is that these pairs might be seen often together in conjunction with another tags.  For example `dataset` and `scikit-learn` might have a strong relation to `pandas`, but each by itself don't.  In a subsequent study, we could examine such relations using a powerful data mining technique: association rules.

## Is Deep Learning Just a Fad?

Before we solidify a recommendation, it is important to make sure that deep learning is not just a fad.  We would like our content to be relevant for a long period of time.  Let's explore this by examining how many deep learning questions there are relative to the total amount of questions per time period.

We will start by reading in all the questions, not merely the questions from 2019 that we read in previously.  We will also use the same technique as before to clean up the tags.

```python
# Read in all questions
all_questions = pd.read_csv('all_questions.csv', parse_dates = ['CreationDate'])

# Clean the tags
all_questions['Tags'] = all_questions['Tags'].str.replace("^<|>$", "").str.split("><")
```

Now we will need to identify the tags that constitute deep learning.  We will use the tags from most_used that relate to deep learning: `lstm`, `cnn`, `scikit-learn`, `tensorflow`, `keras`, `neural-network`, and `deep-learning`.  

Next we will create a function that identifies deep learning tags by assigning a `1` if it is deep learning or a `0` if it is not.

```python
# Deep learning identification function
def tag_deep_learning(tags):
    for tag in tags:
        if tag in ['lstm', 'cnn', 'scikit-learn', 'tensorflow', 'keras', 'neural-network', 'deep-learning']:
            return 1
        else:
            return 0
        
# Apply function over `all_questions`
all_questions['deep_learning'] = all_questions['Tags'].apply(tag_deep_learning)

# Print a sample of 5 rows
all_questions.sample(5)
```

Everything looks as planned.  

The data science technology field is not dynamic enough to track daily, weekly, or monthly, so we will track it quarterly.

We do not have all of the data for the first quarter of 2020, so we will omit those dates.

```python
# Omit data for first quarter of 2020
all_questions = all_questions[all_questions['CreationDate'].dt.year < 2020]
```

Now we will create a column that identifies the quarter in which the question was asked.

```python
# Function to identify quarter
def identify_quarter(datetime):
    year = str(datetime.year)[-2:]
    quarter = str(((datetime.month -1) // 3) + 1)
    return '{y}Q{q}'.format(y= year, q = quarter)

# Apply function to `all_questions`
all_questions['Quarter'] = all_questions['CreationDate'].apply(identify_quarter)

# Print first 5 rows
all_questions.head()  
```

 Finally, we will group by quarter and:
 * Count the number of deep learning questions
 * Count the total number of questions
 * Compute the ratio between the two numbers above
 

```python
# Group by quarter and count
quarterly = all_questions.groupby('Quarter').agg({'deep_learning': ['sum', 'size']})
quarterly.columns = ['DeepLearningQuestions', 'TotalQuestions']
quarterly['DeepLearningRate'] = quarterly['DeepLearningQuestions'] / quarterly['TotalQuestions']

quarterly.reset_index(inplace = True)
quarterly.head(5)
```

Lastly, we will visualize this data with a bar and line graph.

```python
ax1 = quarterly.plot(x="Quarter", y="DeepLearningRate",
                    kind="line", linestyle="-", marker="o", color="orange",
                    figsize=(25,15)
                    )

ax2 = quarterly.plot(x="Quarter", y="TotalQuestions",
                     kind="bar", ax=ax1, secondary_y=True, alpha=0.7, rot=45)

for idx, t in quarterly["TotalQuestions"].iteritems():
    ax2.text(idx, t, str(t), ha="center", va="bottom")
xlims = ax1.get_xlim()

ax1.get_legend().remove()

handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(handles=handles1 + handles2,
           labels=labels1 + labels2,
           loc="upper left", prop={"size": 12})


for ax in (ax1, ax2):
    for where in ("top", "right"):
        ax.spines[where].set_visible(False)
        ax.tick_params(right=False, labelright=False)
```

Examining this graph, it is apparent that deep learning questions exhibit a high growth since the start of Data Science Stack Exchange in 2014.  Now it looks like the rate is plateauing, but there is no evidence to sugggest that interest in deep learning will decrease.  Thus, our recommendation on creating deep learning content still stands.  An example of this type of content would be to create Python content that uses deep learning for classification problems.
